```python
import pandas as pd
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
import time
import csv

sc = SparkContext()
ssc = StreamingContext(sc, 10)
socket_stream = ssc.socketTextStream("127.0.0.1", 5553)
lines = socket_stream.window( 20 )

def dump(tweet):
    with open('tweets.csv', mode='a') as csv_file:
        fieldnames = ['tweet']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writerow({'tweet': tweet})

# create a function dump_json that takes a tweet as input and dumps it to a json file
def dump_json(tweet):
    with open('tweets.json', mode='a') as json_file:
        json_file.write(tweet)

# create a funcion dump_mongo that takes the tweets and save it to a mongo database
import pymongo
from pymongo import MongoClient
def dump_mongo(tweet):
    client = MongoClient('localhost', 27017)
    db = client.tweets
    collection = db.tweets
    collection.insert_one(tweet)


lines.foreachRDD(lambda rdd: 
    rdd.foreach(lambda tweet: dump(tweet))
)


ssc.start()
```
